```python
import numpy as np

# Membuat titik-titik yang sesuai dengan data yang diberikan
A = np.array([40, 43])
B = np.array([101, 1])
C = np.array([111, 54])
D = np.array([104, 65])
E = np.array([60, 22])
F = np.array([20, 2])

# Menggabungkan titik-titik ke dalam matriks
pts = np.vstack((A, B, C, D, E, F))

# Menampilkan matriks titik-titik
print(pts)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.spatial import distance  # Import modul distance

# Data titik-titik
pts = np.array([[40, 43],
                [101, 1],
                [111, 54],
                [104, 65],
                [60, 22],
                [20, 2]])

# Menghitung matriks jarak
dis = distance.cdist(pts, pts, 'euclidean')

# Menampilkan matriks jarak sebagai peta panas
plt.figure(figsize=(8, 6))
sns.heatmap(dis, annot=True, cmap='coolwarm', fmt='.3f', xticklabels=['A', 'B', 'C', 'D', 'E', 'F'], yticklabels=['A', 'B', 'C', 'D', 'E', 'F'])
plt.xlabel('Titik')
plt.ylabel('Titik')
plt.title('Matriks Jarak Antara Titik-titik')
plt.show()
```

```python
import numpy as np

# Matriks kebalikan W yang telah diisi sebelumnya
W = np.array([[np.inf, 0.0135, 0.0139, 0.0148, 0.0345, 0.0219],
              [0.0135, np.inf, 0.0185, 0.0156, 0.0217, 0.0123],
              [0.0139, 0.0185, np.inf, 0.0767, 0.0166, 0.0095],
              [0.0148, 0.0156, 0.0767, np.inf, 0.0163, 0.0095],
              [0.0345, 0.0217, 0.0166, 0.0163, np.inf, 0.0224],
              [0.0219, 0.0123, 0.0095, 0.0095, 0.0224, np.inf]])

# Mengganti nilai tak terhingga dengan NaN
W[~np.isfinite(W)] = np.nan

# Menghitung jumlah baris dengan mengabaikan NaN
rtot = np.nansum(W, axis=1)
print("Jumlah baris setelah mengabaikan NaN:")
print(rtot)

# Membagi matriks W oleh jumlah baris
W = W / rtot[:, np.newaxis]

# Menghitung jumlah baris dan kolom setelah normalisasi
row_sums = np.nansum(W, axis=1)
col_sums = np.nansum(W, axis=0)

print("Jumlah baris setelah normalisasi:")
print(row_sums)
print("Jumlah kolom setelah normalisasi:")
print(col_sums)
```

```python
import libpysal as lps
import geopandas as gpd
import networkx as nx
import matplotlib.pyplot as plt

# Membaca shapefile
p = gpd.read_file(dataset_url)

# Membuat daftar tetangga (neighbor list) dengan metode Queen
wr = lps.weights.Queen.from_dataframe(p)

# Mengonversi daftar tetangga menjadi graf
G = nx.Graph(wr.sparse)

# Menampilkan plot dari shapefile
ax = p.plot(color='gray', edgecolor='blue')

# Menampilkan garis antar titik tetangga
for edge in G.edges:
    start_point = p.geometry.iloc[edge[0]].centroid
    end_point = p.geometry.iloc[edge[1]].centroid
    plt.plot([start_point.x, end_point.x], [start_point.y, end_point.y], color='red', linewidth=1)

plt.show()
```
